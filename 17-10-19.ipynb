### matplotlib
# -pyplot

```python
import matplotlib.pyplot as plt
distance=[1,2,3,4,5]
time=[2,3,4,5,6]
plt.plot(time,distance)
plt.show()
```

```python
import matplotlib.pyplot as plt
distance=[1,2,3,4,5]
time=[2,3,4,5,6]
plt.plot(time,distance,'g*')
plt.xlabel('time')
plt.ylabel('distance')
plt.title("distance vs time plot")
plt.grid()
```

```python
plt.subplot(1,2,1)
plt.plot([1,2,3],[4,5,6])
plt.show()

plt.subplot(1,2,2)
plt.plot([2,3,4],[3,4,2])
plt.show()
```
